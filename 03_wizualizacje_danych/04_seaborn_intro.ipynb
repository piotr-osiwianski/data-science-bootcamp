### <a name='3'></a> Tips data

```
df = sns.load_dataset('tips')
df.head(10)
```

```
df.info()
```

```
# widok transponowany .T
df.describe().T
```

```
df.describe(include=['category']).T
```

```
# wykres relacyjny, calkowity rachunek do napiwku
_ = sns.relplot(data=df, x='total_bill', y='tip')
```

```
# font_scale - zwiekszy czcionke
sns.set(font_scale=1.2)
_ = sns.relplot(data=df, x='total_bill', y='tip')
```

```
# size - rozmiar grupy w restauracji - jako 3 zmienna, hue - kolor
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size')
```

```
# col='time', zmienna kategoryczna ktora nie ma duzych wartosci, podzieli nam na dwa wykresy
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', col='time')
```

```
# row='time' - to samo co poprzednio tylko dane w pionie
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', row='time')
```

```
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', col='time', row='smoker')
```

```
# _ - przypisujemy do zmiennej zeby nie miec info o obiekcie,
# x='day' zmienna kategoryczna ktora przyjmuje dwie wartosci
_ = sns.catplot(data=df, x='day', y='total_bill')
```

```
_ = sns.catplot(data=df, x='day', y='total_bill', kind='swarm')
```

```
_ = sns.catplot(data=df, x='day', y='total_bill', kind='box')
```

```
_ = sns.catplot(data=df, x='day', y='total_bill', kind='violin')
```

```
_ = sns.catplot(data=df, x='day', y='total_bill', kind='bar')
```

```
# wykres czestotliwosci danej kategorii
_ = sns.catplot(data=df, x='day', kind='count')
```

### <a name='4'></a> Titanic data

```
df = sns.load_dataset('titanic')
df.head()
```

```
# rodzaj pokladu i czestotliwosc
sns.catplot(data=df, x='deck', kind='count', palette='Blues')
```

```
# to samo tylko zmieniamy parametr x na y
sns.catplot(data=df, y='deck', kind='count', palette='Blues')
```
